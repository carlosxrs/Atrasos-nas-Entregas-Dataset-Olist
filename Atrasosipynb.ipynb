```python
import pandas as pd

orders = pd.read_csv('olist_orders_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')

orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

orders['tempo_entrega_real'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days
orders['tempo_entrega_estimado'] = (orders['order_estimated_delivery_date'] - orders['order_purchase_timestamp']).dt.days

orders['entrega_atrasada'] = orders['tempo_entrega_real'] > orders['tempo_entrega_estimado']

df = pd.merge(orders, customers[['customer_id', 'customer_state']], on='customer_id', how='left')
```

```python
df_grouped = df.groupby('customer_state').agg(
    total_pedidos=('order_id', 'count'),
    pedidos_atrasados=('entrega_atrasada', 'sum')
).reset_index()

df_grouped['percentual_atraso'] = (df_grouped['pedidos_atrasados'] / df_grouped['total_pedidos']) * 100

df_grouped.to_csv('pedidos_por_estado.csv', index=False, sep=';')
```

```python
import pandas as pd

df = pd.read_csv('/content/olist_orders_dataset.csv', sep=',')

df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'], errors='coerce')
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'], errors='coerce')

df['entrega_atrasada'] = df['order_delivered_customer_date'] > df['order_estimated_delivery_date']

df_customers = pd.read_csv('/content/olist_customers_dataset.csv', sep=',')

df = df.merge(df_customers[['customer_id', 'customer_state']], on='customer_id', how='left')


df_grouped = df.groupby('customer_state').agg(
    total_pedidos=('order_id', 'count'),
    pedidos_atrasados=('entrega_atrasada', 'sum')
)


df_grouped['percentual_atraso'] = (df_grouped['pedidos_atrasados'] / df_grouped['total_pedidos']) * 100


df_grouped_reset = df_grouped.reset_index()


df_grouped_reset['percentual_atraso'] = df_grouped_reset['percentual_atraso'].round(2)

df_grouped_reset.to_csv('pedidos_por_estado.csv', index=False, sep=';', decimal=',')


df_grouped_reset


```
